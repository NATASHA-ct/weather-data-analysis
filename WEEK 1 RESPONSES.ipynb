```python
# Weather Data Analysis
#This notebook contains the analysis of the weather dataset to answer specific questions.
```

```python
import pandas as pd

# Load the dataset
file_path = '1. Weather Data.csv'
df = pd.read_csv(file_path)
```

## Question 1
Find all records where the weather was exactly clear.


```python
clear_weather_records = df[df['Weather_Condition'] == 'Clear']
clear_weather_records
```

## Question 2
Find the number of times the wind speed was exactly 4 km/hr.


```python
wind_speed_count = df[df['Wind Speed_km/h'] == 4].shape[0]
wind_speed_count
```

## Question 3
Check if there are any NULL values present in the dataset


```python
null_values = df.isnull().sum()
null_values
```

## Question 4
Rename the column "Weather" to "Weather_Condition."

```python
df.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
df.head()

```

## Question 5
What is the mean visibility of the dataset?

```python
mean_visibility = df['Visibility_km'].mean()
mean_visibility
```

# Question 6
Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.y


```python
records_count = df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)].shape[0]
records_count
```

# Question 7
What is the mean value of each column for each weather condition?

```python
mean_values_by_weather = df.groupby('Weather_Condition').mean()
mean_values_by_weather
```

# Question 8
Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.

```python
filtered_records = df[(df['Weather_Condition'] == 'Clear') & ((df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40))]
filtered_records
```

# Question 9
Find the number of weather conditions that include snow.

```python

snow_conditions_count = df['Weather_Condition'].str.contains('Snow', case=False).sum()
snow_conditions_count
```
